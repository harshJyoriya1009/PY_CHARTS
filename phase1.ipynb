```python
x=['Mon','Tue','Wed','Thu','Fri']
y=[24,56,15,44,34]

plt.plot(x,y)
plt.title("Shoes Selling")
plt.xlabel("Days")
plt.ylabel("Number of Shoes")
plt.grid(True)
plt.show()
```
