```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,25,40,45]

plt.plot(x,y)
plt.show()
```

```python
x=['Mon','Tue','Wed','Thu','Fri']
y=[24,56,15,44,34]

plt.plot(x,y)
plt.title("Shoes Selling")
plt.xlabel("Days")
plt.ylabel("Number of Shoes")
plt.grid(True)
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May']
sales=[140,120,130,200,150]

plt.plot(months, sales, color="green", linestyle='-.', linewidth=2, marker='o', label='2024 Sales Data')
plt.xlabel("Months")
plt.ylabel("Number of Phones")
plt.title("Phone Selling Data of 5 Monthhs")
plt.legend(loc='upper left', fontsize=10)
plt.grid(color='gray', linestyle='-.', linewidth=1)
plt.xlim(0,4)
plt.ylim(100,250)
plt.show()
```

```python
x=['A','B','C','D']
y=[1200,1300,1250,1400]

plt.bar(x, y, color='red', label='Sales 2025')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Total sales per month")
plt.legend()
plt.show()
```

### Pie Charts

```python
# plt.pie(values, labels=label_list, color=color_list autopct='%1.1f%%')

# autopct means auto percentage which generally add percentage sign and also add point value in pie chart
```

```python
regions=['East', 'West', 'North', "South"]
revenue=[1000, 2300, 4500, 1500]

plt.pie(revenue, labels=regions, autopct='%1.1f%%',  colors=['red', 'blue', "gray", 'gold'])
plt.title("Revenue collection")
plt.show()
```

```python
# plt.hist(data, bins=no_of_bins, color="color_name", edgecolor='border_color')

# bins generally means divide data with how many range  

# Height of bars represent no.of students in that range
```

```python
scores=[45,76,98,23,67,67,60,75,65,42,33,78,58,26,58,65,35,75,65,50,69]

plt.hist(scores, bins=5, color="gold", edgecolor="black")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Score distribution of Students")
plt.show()
```
