```python
import matplotlib.pyplot as plt
```

### Scatter plot

```python
# plt.scatter(x, y, color='color name', marker='marker style', label='label name')
```

```python
hrs_studed=[1,2,3,4,5,6,7,8]
exam_score=[25,35,45,65,78,98,99,100]

plt.scatter(hrs_studed, exam_score, color='gold', marker='o', label='Student Data')
plt.xlabel("Numer of hours student studied")
plt.ylabel("Marks of the Student")
plt.legend()
plt.title("Relationship between Study time and Exam Score")
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3], [65,72,98], color='yellow', marker='o', label='Section A')
plt.scatter([1,2,3], [78,71,90], color='blue', marker='^', label='Section B')

plt.xlabel("Numer of hours student studied")
plt.ylabel("Marks of the Student")
plt.legend()
plt.title("Comparison of Two classes")
plt.grid()
plt.show()
```

# Subplots

```python
'''
1) Multiple Charts
2) Used less Space
3) Quickly

Mainly using in Dashboard
'''

# plt.subplot(nrows, ncols, index)
```

```python
x=[1,2,3,4,5]
y=[10,12,15,18,20]

plt.subplot(1,2,1) # 1rows, 2 column, 1 position
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1,2,2) # 1 rows, 2 column, 2 position
plt.bar(x,y)
plt.title("Bar chart")

plt.show()
```

```python
# It is more proffesional and clear way for big code..
# fig, ax=plt.subplots(nrows, ncols, figsize=(width, height))   

fig, ax= plt.subplots(1, 2, figsize=(10,5))

x=[1,2,3,4]
y=[10,20,15,30]

ax[0].plot(x,y, color="red")
ax[0].set_title("Line plot")

ax[1].bar(x,y,color="green")
ax[1].set_title("Bar chart")

plt.suptitle("Comparison between Line and Bar chart")
plt.tight_layout()
plt.show()
```
